```python
import numpy as np 
import os
import cv2
import pickle
import matplotlib.pyplot as plt
```

```python
file_path0='healthy_skin'
file_path1='Skin cancer ISIC The International Skin Imaging Collaboration/images'
file_path2='nevus'
images0=[]
images1=[]
images2=[]

for filename in os.listdir(file_path0):
    image_path=os.path.join(file_path0, filename)
    img=cv2.imread(image_path)
    if img is not None:
        resize=cv2.resize(img, (28, 28), interpolation=cv2.INTER_AREA)
        gray=cv2.cvtColor(resize, cv2.COLOR_BGR2GRAY)
        normalized=gray.astype('float32')/255.0
        images0.append(normalized)


for filename in os.listdir(file_path1):
    image_path=os.path.join(file_path1, filename)
    img=cv2.imread(image_path)
    if img is not None:
        resize=cv2.resize(img, (28, 28), interpolation=cv2.INTER_AREA)
        gray=cv2.cvtColor(resize, cv2.COLOR_BGR2GRAY)
        normalized=gray.astype('float32')/255.0
        images1.append(normalized)



for filename in os.listdir(file_path2):
    image_path=os.path.join(file_path2, filename)
    img=cv2.imread(image_path)
    if img is not None:
        resize=cv2.resize(img, (28, 28), interpolation=cv2.INTER_AREA)
        gray=cv2.cvtColor(resize, cv2.COLOR_BGR2GRAY)
        normalized=gray.astype('float32')/255.0
        images2.append(normalized)




y0=[0]*len(images0)
y1=[1]*len(images1)
y2=[2]*len(images2)


images=np.array(images0+images1+images2)
label=np.array(y0+y1+y2)


print(len(images))
print(len(label))
```

```python
print(np.unique(label))
```

```python
data={'processed_image_data':images, 'processed_class_data':label}
with open('dataset.pkl', 'wb') as file:
    pickle.dump(data, file)
    print('dataset saved in path dataset.pkl')
```
